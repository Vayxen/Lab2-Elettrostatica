```python
#istogramma Q
import numpy as np
import matplotlib.pyplot as plt

Qdata = np.array([3.2, 3.4, 3.1, 3.3, 3.3, 3.4, 3.3,3.0,3.1,2.8,3.2,2.5,2.5,2.8,3.0,2.8,2.8,2.8,2.8,3.0,3.0,2.9,3.1,3.2,2.8,3.1,3.3,3.4,3.3,3.2,3.5,3.4,3.6,3.5,3.3,3.4,3.3,3.2,3.0,3.4, 3.2, 3.1, 3.3, 3.2, 3.2, 3.2, 3.3, 3.1, 3.3])
n, bins, patches = plt.hist(Qdata, bins=5,density=True)

from scipy.stats import norm
(mu, sigma) = norm.fit(Qdata)
plt.title(r'$\mathrm{Histogram\ of\ Q:}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))

pdf_x = np.linspace(np.min(Qdata),np.max(Qdata),100)
pdf_y = 1.0//sigma/np.sqrt(2*np.pi)*np.exp(-0.5*(pdf_x-mu)**2/2/sigma**2)
plt.plot(pdf_x,pdf_y,'k--')

plt.grid(color='lightgray',linestyle="--",zorder=0)
plt.minorticks_on()
plt.grid(True,"minor",linestyle="--",color='lightgray')
plt.show()
```
